#### Descripción de la base de datos
Alcohol en estudiantes: Datos del consumo de alcohol de estudiantes de diferentes cursos.
Attributes for both student-mat.csv (Math course) and student-por.csv (Portuguese language course) datasets:
- school - student's school (binary: 'GP' - Gabriel Pereira or 'MS' - Mousinho da Silveira)
- sex - student's sex (binary: 'F' - female or 'M' - male)
- age - student's age (numeric: from 15 to 22)
- address - student's home address type (binary: 'U' - urban or 'R' - rural)
- famsize - family size (binary: 'LE3' - less or equal to 3 or 'GT3' - greater than 3)
- Pstatus - parent's cohabitation status (binary: 'T' - living together or 'A' - apart)
- Medu - mother's education (numeric: 0 - none, 1 - primary education (4th grade), 2 – 5th to 9th grade, 3 – secondary education or 4 – higher education)
- Fedu - father's education (numeric: 0 - none, 1 - primary education (4th grade), 2 – 5th to 9th grade, 3 – secondary education or 4 – higher education)
- Mjob - mother's job (nominal: 'teacher', 'health' care related, civil 'services' (e.g. administrative or police), 'at_home' or 'other')
- Fjob - father's job (nominal: 'teacher', 'health' care related, civil 'services' (e.g. administrative or police), 'at_home' or 'other')
- reason - reason to choose this school (nominal: close to 'home', school 'reputation', 'course' preference or 'other')
- guardian - student's guardian (nominal: 'mother', 'father' or 'other')
- traveltime - home to school travel time (numeric: 1 - <15 min., 2 - 15 to 30 min., 3 - 30 min. to 1 hour, or 4 - >1 hour)
- studytime - weekly study time (numeric: 1 - <2 hours, 2 - 2 to 5 hours, 3 - 5 to 10 hours, or 4 - >10 hours)
- failures - number of past class failures (numeric: n if 1<=n<3, else 4)
- schoolsup - extra educational support (binary: yes or no)
- famsup - family educational support (binary: yes or no)
- paid - extra paid classes within the course subject (Math or Portuguese) (binary: yes or no)
- activities - extra-curricular activities (binary: yes or no)
- nursery - attended nursery school (binary: yes or no)
- higher - wants to take higher education (binary: yes or no)
- internet - Internet access at home (binary: yes or no)
- romantic - with a romantic relationship (binary: yes or no)
- famrel - quality of family relationships (numeric: from 1 - very bad to 5 - excellent)
- freetime - free time after school (numeric: from 1 - very low to 5 - very high)
- goout - going out with friends (numeric: from 1 - very low to 5 - very high)
- Dalc - workday alcohol consumption (numeric: from 1 - very low to 5 - very high)
- Walc - weekend alcohol consumption (numeric: from 1 - very low to 5 - very high)
- health - current health status (numeric: from 1 - very bad to 5 - very good)
- absences - number of school absences (numeric: from 0 to 93)
  
These grades are related with the course subject, Math or Portuguese:

- G1 - first period grade (numeric: from 0 to 20)
- G2 - second period grade (numeric: from 0 to 20)
- G3 - final grade (numeric: from 0 to 20, output target)

##### Consumo de alcohol vs Edad

```python
#Agrupar valores de estudiantes que no 
count_por_grupo_walc= df4[df4['Walc']>2].groupby('age')['Walc'].count()
count_por_grupo_dalc = df4[df4['Dalc']>2].groupby('age')['Dalc'].count()

plt.figure(figsize=(10, 6))
plt.rc('font', size=8)

# Crear un DataFrame con los datos
data = pd.DataFrame({
    'Fin de semana': count_por_grupo_walc,
    'Entre semana': count_por_grupo_dalc
})
data.plot(kind='bar', width=0.8)
plt.title('Consumo de alcohol vs Edad')
plt.xlabel('Edad')
plt.ylabel('Consumo de alcohol')
plt.legend()
plt.xticks(rotation=0) 
plt.show()
```

Este análisis sugiere que el consumo de alcohol está fuertemente influenciado por la edad, con un aumento en los niveles de consumo a medida que los estudiantes se acercan a la mayoría de edad. Además, los estudiantes tienden a consumir más alcohol durante los fines de semana que entre semana, lo cual puede estar relacionado con la disponibilidad de tiempo libre y eventos sociales.

##### Consumo de alcohol vs Estado de Salud

```python
mapeo = {
    1: '1: Muy mal',
    2: '2: Mal',
    3: '3: Regular',
    4: '4: Bueno',
    5: '5: Muy bueno'
}
df4['health1'] = df4['health'].map(mapeo)
estado_de_salud_walc = df4[df4['Walc']>2].groupby('health1')['Walc'].count()
estado_de_salud_dalc = df4[df4['Dalc']>2].groupby('health1')['Dalc'].count()
plt.figure(figsize=(10, 6))
plt.rc('font', size=8)

# Crear un DataFrame con los datos
data = pd.DataFrame({
    'Fin de semana': estado_de_salud_walc,
    'Entre semana': estado_de_salud_dalc
})
data.plot(kind='bar', width=0.8)
plt.title('Consumo de alcohol vs Estado de Salud')
plt.xlabel('Estado de Salud')
plt.ylabel('Consumo de alcohol')
plt.xticks(rotation=0) 
plt.legend()
plt.show()
```

Los estudiantes con mejor estado de salud percibido tienden a consumir más alcohol, especialmente los fines de semana.
El consumo de alcohol es mayor los fines de semana en todos los grupos de salud, pero es más pronunciado en aquellos que reportan buena o muy buena salud.
Los estudiantes con peor salud percibida consumen menos alcohol, tanto entre semana como los fines de semana.
En resumen, un mejor estado de salud está asociado con un mayor consumo de alcohol, particularmente en momentos sociales como los fines de semana.

##### Consumo de alcohol (Walc) según Estado de Salud (health1)

```python
df_res=pd.DataFrame(df4.groupby(by=['health1','Walc'])['Walc'].count())
df_res=df_res.rename(columns={"Walc":"counts"})
df_res=df_res.reset_index()
df_res
mapeo = {
    1: '1: Muy poco',
    2: '2: Poco',
    3: '3: Moderado',
    4: '4: Varias Veces',
    5: '5: Muchas Veces'
}
df_res['Walc']=df_res.Walc.map(mapeo)
df_pivot = df_res.pivot(index='health1', columns='Walc', values='counts')
plt.figure(figsize=(12, 8))
df_pivot.plot(kind='bar', width=0.8)
plt.title('Consumo de alcohol (Walc) según Estado de Salud (health1)')
plt.xlabel('Estado de Salud')
plt.ylabel('Número de personas')
plt.legend(title='Consumo de alcohol')
plt.xticks(rotation=0)  # Rotar etiquetas del eje X si es necesario
plt.show()
```

El análisis muestra una relación clara entre el estado de salud y la frecuencia de consumo de alcohol. Los datos sugieren que las personas con un mejor estado de salud tienden a reportar un menor consumo de alcohol, mientras que aquellas con peor salud muestran una mayor frecuencia de consumo. Específicamente, se observa que los grupos con mejores niveles de salud (health1) tienen más individuos en las categorías de consumo bajo o moderado, mientras que el consumo más frecuente (varias veces o muchas veces) se concentra en grupos con peor salud. Estos hallazgos sugieren una posible asociación negativa entre el consumo de alcohol y el estado general de salud, destacando la importancia de considerar los hábitos de consumo como un factor relevante en la evaluación y promoción de la salud.

##### Consumo de alcohol entre semana (Dalc) según Estado de Salud (health1)

```python
df_res=pd.DataFrame(df4.groupby(by=['health1','Dalc'])['Dalc'].count())
df_res=df_res.rename(columns={"Dalc":"counts"})
df_res=df_res.reset_index()
mapeo = {
    1: '1: Muy poco',
    2: '2: Poco',
    3: '3: Moderado',
    4: '4: Varias Veces',
    5: '5: Muchas Veces'
}
df_res['Dalc']=df_res.Dalc.map(mapeo)
df_pivot = df_res.pivot(index='health1', columns='Dalc', values='counts')
plt.figure(figsize=(12, 8))
df_pivot.plot(kind='bar', width=0.8)
plt.title('Consumo de alcohol entre semana (Dalc) según Estado de Salud (health1)')
plt.xlabel('Estado de Salud')
plt.ylabel('Número de personas')
plt.legend(title='Consumo de alcohol')
plt.xticks(rotation=0)  # Rotar etiquetas del eje X si es necesario
plt.show()
```

El gráfico de barras sugiere que las personas con mejor estado de salud tienden a consumir menos alcohol entre semana, mostrando frecuencias más altas en las categorías de consumo "Muy poco" o "Poco". En contraste, aquellos con un estado de salud más bajo tienden a estar más representados en las categorías de consumo más frecuente ("Varias veces" o "Muchas veces"). Este patrón indica una asociación entre un menor consumo de alcohol durante los días de semana y un mejor estado de salud, destacando la influencia potencial del consumo regular de alcohol sobre el bienestar general de las personas

En resumen, ambos gráficos subrayan la relación negativa entre el consumo de alcohol y el estado de salud, pero también destacan que el consumo excesivo durante los días laborables puede ser particularmente perjudicial para la salud general.

##### Calificacion por Periodo (G1, G2, G3) y Materia vs Consumo de alcohol

```python
sns.set(style="whitegrid")

# nombre de las columnas for G1, G2, G3
grade_columns = ['G1', 'G2', 'G3']

#filtramos por alumnos que beban entre semana:
df_dalc = df4[df4['Dalc']>2]

# se crea figura con 3 axes vertigales
fig, axes = plt.subplots(1, 3, figsize=(20, 10))


# se crea un bucle para poder graficar cada periodo de calificaciones
for i, grade in enumerate(grade_columns):
    sns.boxplot(ax=axes[i], x='class', y=grade, data=df_dalc)
    axes[i].set_title(f'{grade} Calificacion por Clase Dalc')
    axes[i].set_ylabel(f'{grade} Grades')
    axes[i].set_xlabel('Class')

df_walc = df4[df4['Walc']>2]
# se crea figura con 3 axes vertigales
fig, axes = plt.subplots(1, 3, figsize=(20, 10))
# se crea un bucle para poder graficar cada periodo de calificaciones
for i, grade in enumerate(grade_columns):
    sns.boxplot(ax=axes[i], x='class', y=grade, data=df_walc)
    axes[i].set_title(f'{grade} Calificacion por Clase Walc')
    axes[i].set_ylabel(f'{grade} Grades')
    axes[i].set_xlabel('Class')
# Adjust the layout to avoid overlap
plt.tight_layout()
plt.show()
```

El consumo elevado de alcohol, especialmente durante la semana, está vinculado a un menor rendimiento académico, reflejado en calificaciones más bajas y una mayor variabilidad en los resultados.
